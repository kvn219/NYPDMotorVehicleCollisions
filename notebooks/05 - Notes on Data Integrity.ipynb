```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import missingno as msno

import seaborn as sns

# Pandas Printing Options
pd.set_option('display.float_format', lambda x: '%.3f' % x)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

sns.set_style('white')
```

```python
data = pd.read_csv("../data/nypd_collision_data.csv", low_memory=True)
```

# VISUALIZE MISSING DATA

```python
msno.matrix(data)
```

```python
data['DATE'] = pd.to_datetime(data['DATE'] + ' ' + data['TIME'])
```

```python
data = data.set_index('DATE')
```

# SPECIFIED

```python
specified = data[data['CONTRIBUTING FACTOR VEHICLE 1'] != 'Unspecified']
specified.shape
```

```python
plt.figure(figsize=(16, 6))
plt.plot(specified.resample('m')['UNIQUE KEY'].count(), label='Specified')
plt.ylabel('# of Collisions')
plt.legend();
```

```python
specified.groupby('BOROUGH').resample('m')['UNIQUE KEY'].count().unstack().T.plot(figsize=(16, 6));
```

# UNSPECIFIED

```python
unspecified = data[data['CONTRIBUTING FACTOR VEHICLE 1'] == 'Unspecified']
unspecified.shape
```

```python
plt.figure(figsize=(16, 6))
plt.plot(unspecified.resample('m')['UNIQUE KEY'].count(), label='Unspecified')
plt.ylabel('# of Collisions')
plt.legend();
```

```python
unspecified.groupby('BOROUGH').resample('m')['UNIQUE KEY'].count().unstack().T.plot(figsize=(16, 6));
```

# THE SPECIFICATION MYSTERY

```python
plt.figure(figsize=(16, 6))
plt.plot(unspecified.resample('m')['UNIQUE KEY'].count(), label='Unspecified')
plt.plot(specified.resample('m')['UNIQUE KEY'].count(), label='Specified')
plt.ylabel('# of Collisions')
plt.legend();
```

```python
plt.figure(figsize=(16, 6))
plt.plot(unspecified.resample('m')['UNIQUE KEY'].count(), label='Unspecified')
plt.plot(specified.resample('m')['UNIQUE KEY'].count(), label='Specified')
plt.plot(data.resample('m')['UNIQUE KEY'].count(), label='Total')
plt.ylabel('# of Collisions')
plt.legend();
```
